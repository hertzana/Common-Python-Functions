## Scatterplots

```python
# set graph background to white grid
sns.set_style("whitegrid")
# Scatterplot between age and fare purchased
sns.jointplot(x='Age', y='Fare', data=t)
```

```python
# Plotting quantitative data against categorical variable
# What is fare for at each point of embarkation?
sns.stripplot(x='Embarked', y='Fare', data=t)
```

```python
# For scatter plot above, separate out overlapping data points
sns.stripplot(x='Embarked', y='Fare', data=t, hue='Pclass', jitter=True)
```

```python
# Do the same as above with swarmplot
sns.swarmplot(x='Embarked', y='Fare', data=t)
```

```python
# Separate out data by Pclass, and plot data horizontally
# Below, color data points by passenger class
sns.swarmplot(x='Fare', y='Embarked`', hue='Pclass', data=t)
```
